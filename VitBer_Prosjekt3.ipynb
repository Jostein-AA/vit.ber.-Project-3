# Oppgave 1 a)

Skal vise at strukturligningen:

$\frac{1}{r^{2}}\frac{d}{dr}\left(\frac{r^{2}}{\theta}\frac{dP}{dr}\right)=-\frac{4\pi G\epsilon(r)}{c^{4}}$,

kan skrives om til:

$\frac{1}{\xi^{2}}\frac{d}{d\xi}\xi^{2}\frac{d\theta}{d\xi}=-\theta^{n}$.

Vi har tilstandsligning for $P$, gitt ved:

$P=K\epsilon^{\gamma}$.

$\gamma = 1 + \frac{1}{n}$.

Vi innfører:

$\epsilon = \epsilon_{0}\theta^{n}$

og

$r=a\xi$,

hvor $a$ er gitt ved:

$a=\left[\frac{(n+1)Kc^{4}\epsilon_{0}^{\frac{1}{n}-1}}{4\pi G}\right]^{\frac{1}{2}}$.

Vi får at: $\frac{1}{a}\frac{d}{d\xi}=\frac{d}{dr}$.

Vi setter inn med de nye variablene:

$\frac{1}{a^{2}\xi^{2}}\left(\frac{1}{a}\frac{d}{d\xi}\right)\left(\frac{a^{2}\xi^{2}}{\epsilon_{0}\theta^{n}}\frac{dK\left(\epsilon_{0}\theta^{n}\right)^{\gamma}}{ad\xi}\right)=-\frac{4\pi G\epsilon_{0}\theta^{n}}{c^{4}}$.

Ved å bytte om på konstanter får en:

$\frac{1}{\xi^{2}}\frac{d}{d\xi}\left(\frac{\xi^{2}}{\theta^{n}}\frac{d\theta^{n+1}}{d\xi}\right)=\frac{a^{2}}{K\epsilon_{0}^{\frac{1}{n}}}\left(-\frac{4\pi G\epsilon_{0}\theta^{n}}{c^{4}}\right)$, 

og får da:

$\frac{1}{\xi^{2}}\frac{d}{d\xi}\left(\frac{\xi^{2}}{\theta^{n}}\frac{d\theta^{n+1}}{d\xi}\right)=-(n+1)\theta^{n}$.


Har fra kjerneregelen at:

$\frac{dx^{n+1}}{dr}=(n+1)x^{n}\frac{dx}{dr}.$

Dette medfører at

$\frac{1}{\xi^{2}}\frac{d}{d\xi}\left(\frac{\xi^{2}}{\theta^{n}}\frac{d\theta^{n+1}}{d\xi}\right)=(n+1)\frac{1}{\xi^{2}}\frac{d}{d\xi}\left(\xi^{2}\frac{d\theta}{d\xi}\right),$

noe som gir:

$(n+1)\frac{1}{\xi^{2}}\frac{d}{d\xi}\left(\xi^{2}\frac{d\theta}{d\xi}\right)=-(n+1)\theta^{n}.$


Vi får at


$\frac{1}{\xi^{2}}\frac{d}{d\xi}\left(\xi^{2}\frac{d\theta}{d\xi}\right)=-\theta^{n}$.

Q.E.D.

Likningen ovenfor kalles for Lane-Emden ligningen.

# Oppgave 1 c)

Skal løse Lane-Emden ligningen for $n=1$ med substitusjonen $\theta(\xi) =\frac{u(\xi)}{\xi}$.

$\frac{1}{\xi^{2}}\frac{d}{d\xi}(\xi^{2}\frac{d\theta}{d\xi})=\theta=\frac{u}{\xi}$

Med dette følger  

$\frac{d\theta}{d\xi}=\frac{d}{d\xi}(\frac{u(\xi)}{\xi})=\frac{\xi u'(\xi)-u(\xi)}{\xi^{2}},$  

som vi så setter inn:

$\frac{d}{d\xi}(\xi^{2}\frac{d\theta}{d\xi})=\frac{d}{d\xi}(\xi u'(\xi)-u(\xi))=u'(\xi)+\xi u''(\xi)-u'(\xi)=\xi u''(\xi).$

Setter inn i Lane-Emden ligningen som nå ser slik ut:

$\frac{1}{\xi^{2}}\xi u''(\xi)=-\frac{u(\xi)}{\xi}.$

Forenklet ser vi at uttrykket ligner på en harmonisk oscillator:  

$u''(\xi)=-u(\xi).$

Vi antar derfor en løsning på formen:

$u(\xi)=A\sin(\xi)+B\cos(\xi).$

Substituerer tilbake $\theta=\frac{u(\xi)}{\xi}$ og får:

$\theta(\xi)=\frac{A\sin(\xi)}{\xi}+\frac{b\cos(\xi)}{\xi}.$

Bruker så randbetingelsen $\theta(0)=1$ til å bestemme $B=0$, og $A=1$.  Den andre randbetingelsen blir automatisk oppfylt dersom man deriverer og sjekker. 

Dette gir løsningen:

$\theta(\xi)=\frac{\sin(\xi)}{\xi}.$

# Oppgave 2 a)

Vi innfører:

$\bar{P}=\frac{P}{\rho_{0}c^{2}}$,

$x=\frac{r}{R}$,

og

$\alpha=\frac{r_{s}}{R}$.

Vi har at:

$r_{s}=\frac{2GM}{c^{2}},$

$M=\frac{4\pi\rho_{0}R^{3}}{3},$

og

$\frac{1}{R}\frac{d}{dx}=\frac{d}{dr},$

slik at $\frac{dP}{dr}=\frac{\rho_{0}c^{2}}{R}\frac{d\bar{P}}{dx}$

Skal vise at:

$\frac{d\bar{P}}{dx}=-\frac{1}{2}\alpha x\left[1+\bar{P}\right]\left[1+3\bar{P}\right]\left[1-x^{2}\alpha\right]^{-1}.$

Starter med:


$\frac{dP}{dr}=-\frac{G\epsilon(r)m(r)}{c^{2}r^{2}}\left[1+\frac{P}{\epsilon(r)}\right]\left[1+\frac{4\pi r^{3}P}{mc^{2}}\right]\left[1-\frac{2Gm}{c^{2}r}\right]^{-1}.$


Ettersom $c^2$ og $\rho_{0}$ er uavhengig $r$, så kan $P=\bar{P}c^{2}\rho_{0}$ settes inn, noe som gir

$\frac{d\bar{P}}{dr}=-\frac{G\epsilon(r)m(r)}{c^{2}r^{2}}\left[1+\frac{\bar{P}c^{2}\rho_{0}}{\epsilon(r)}\right]\left[1+\frac{4\pi r^{3}\bar{P}c^{2}\rho_{0}}{mc^{2}}\right]\left[1-\frac{2Gm}{c^{2}r}\right]^{-1}.$

Benytter $\frac{dP}{dr}=\frac{\rho_{0}c^{2}}{R}\frac{d\bar{P}}{dx}$ og får:

$\frac{d\bar{P}}{dx}=\frac{R}{\rho_{0}c^{2}}\frac{dP}{dr}=-\frac{RG\epsilon(r)m(r)}{c^{4}\rho_{0}r^{2}}\left[1+\frac{\bar{P}c^{2}\rho_{0}}{\epsilon(r)}\right]\left[1+\frac{4\pi r^{3}\bar{P}c^{2}\rho_{0}}{mc^{2}}\right]\left[1-\frac{2Gm}{c^{2}r}\right]^{-1}.$

Siden $\rho$ er konstant, vil $\epsilon(r)=\rho_{0}c^{2}$ og $m(r)=\frac{4\pi r^{3}\rho_{0}}{3}$, hvilket gir:

$\frac{d\bar{P}}{dx}=-\frac{RG}{c^{2}r^{2}}\frac{4\pi r^{3}}{3}\left[1+\bar{P}\right]\left[1+3\bar{P}\right]\left[1-\frac{8G\pi r^{3}}{3c^{2}r}\right]^{-1}.$

Setter inn $r=Rx$ og $M=\frac{4\pi\rho_{0}R^{3}}{3}$, som gir:

$\frac{d\bar{P}}{dx}=-\frac{r_{s}}{2R}x\left[1+\bar{P}\right]\left[1+3\bar{P}\right]\left[1-x^{2}\frac{r_{s}}{R}\right]^{-1}.$

Setter så inn $\alpha=\frac{r_{s}}{R}$ som gir oss det vi skulle finne:

$\frac{d\bar{P}}{dx}=-\frac{1}{2}\alpha x\left[1+\bar{P}\right]\left[1+3\bar{P}\right]\left[1-x^{2}\alpha\right]^{-1}.$

Q.E.D.

# Oppgave 2 b)

Uttrykket som skal løses analytisk er:
$\frac{d\bar{P}}{dx}=-\frac{1}{2}\alpha x\left[1+\bar{P}\right]\left[1+3\bar{P}\right]\left[1-x^{2}\alpha\right]^{-1}.$  

Med separasjon av variable får en utrykket:  

$\frac{d\bar{P}}{1+4\bar{P}+3\bar{P}^{2}}=-\frac{1}{2}\frac{\alpha xdx}{1-x^{2}\alpha}.$  

Integrerer venstre og høyre side og får:  

$\frac{1}{2}\left[ln\left(|3\bar{P}+1|\right)-ln\left(|\bar{P}+1|\right)+\right]=\frac{1}{4}ln\left(|\alpha x^{2}-1|\right)+C,$  

der $C$ er en konstant. Videre opphøyes eulertallet i venstre og høyre side av ligningen, og en får:  

$\frac{\sqrt{|3\bar{P}+1|}}{\sqrt{|\bar{P}+1|}} = C\sqrt[4]{|\alpha x^2-1|}.$  

Siden $\alpha,x\in[0,1]$, og $P<0$ er ufysisk, kan ligningen omskrives til:  

$\frac{\sqrt{3\bar{P}+1}}{\sqrt{\bar{P}+1}} = C\sqrt[4]{1-\alpha x^2}.$  

Etter litt algebra kommer en frem til uttrykket:  

$\bar{P} = \frac{C-\sqrt{1-\alpha x^2}}{\sqrt{1-\alpha x^2}-3C}.$  

Ved bruk av initialverdiproblemet $\bar{P}(1)=0$, finner en at $C=\sqrt{1-\alpha}$. Innsatt i ligningen over gir dette:  

$\bar{P} = \frac{\sqrt{1-\alpha}-\sqrt{1-\alpha x^2}}{\sqrt{1-\alpha x^2}-3\sqrt{1-\alpha}}.$

Q.E.D.

$\bar{P}\approx\frac{1}{4}\alpha(1-x^{2})$.

$y_1=y(x_0)+hf(y_0,x_0)$.  

Vi taylorutvikler $y(x)$ rundt $x=x_0$, og får et

$y(x_0+h)=y_0 + hf(y_0,x_0) + \frac{1}{2}h^2y''(\xi)$,  

der $\xi\in [x_0,x_0+h]$. Avviket mellom korrekt løsning og approksimert løsning er da gitt ved

$\tau_1=y(x_0+h)-y_1 = y_0 + hf(y_0,x_0) + \frac{1}{2}h^2y''(\xi) - y(x_0)-hf(y_0,x_0) = \frac{1}{2}h^2y''(\xi),$  

og en ser at $\tau_1 \propto h^2$.

Feilen i det første steget vil ha tilsvarende uttrykk som den lokale feilen i hvert steg, noe som kan sees av utledningen i a). Vi har altså at $\tau_{local} = \mathcal{O}(h^2)$. Den globale feilen vil være akkumulasjonen av de lokale feilene. Vi kan omskrive $\tau_{local}$ til:  

$\tau_{local} = ah^2,$

$N=\frac{x_N-x_0}{h}.$

**c)**  
Innfører $\chi=\theta'(\xi)$. Ønsker å få Lane-Emden ligningen på formen  

$\begin{bmatrix} \theta'(\xi)\\ \chi'(\xi) \end{bmatrix} = f\left( \begin{bmatrix} \theta \\ \chi \end{bmatrix} \right)$
$= \begin{bmatrix} g_1(\chi) \\ g_2(\theta,\chi,\xi) \end{bmatrix}$.  

Produktregelen i derivasjon benyttes og Lane-Emden ligningen omskrives til formen  

$\chi'(\xi) = \frac{d^2\theta}{d\xi^2} = -\frac{2}{\xi} \frac{d\theta}{d\xi}-\theta^n$.  

En kan da uttrykke Lane-Embden ligningen på ønsket form:  

$f\left(\begin{bmatrix} \theta \\ \chi \end{bmatrix} \right) = \begin{bmatrix} \chi \\ -\frac{2}{\xi} \chi-\theta^n \end{bmatrix}.$  
  

**d)**

```python
import numpy as np
import matplotlib.pyplot as plt

#----------------------------------------------------------------
def LaneEmden(y,xi,n):
    #y=np.array([theta, xi])
    answer = np.array([y[1], -2/xi * y[1] - y[0]**n])
    return answer
    
#-------------------------------------------------------------
#y0 er en np.array(theta0,chi0) (Randbetingelsene)
#xi0=0, theta(xiEnd)=0
def eulersmethod(y0, h, f, n):
    y=[y0]; xi=[10**-3] #starter med liten xi0 siden 0/0 blir tull
    
    while True:
        
        S=f(y[-1],xi[-1],n)
        y_iPlussEn=y[-1]+h*S
        
        #y-iPlussEn er siste utregnede theta. Denne er positiv helt til vi er ved enden av planeten.
        #Valgte denne if-løsningen siden RK4 skaper pythonproblemer når theta og chi er veldig små, og unngår dermed
        #feilkode med "nan". Ønsker å være konsekvent, og bruker derfor denne løsningen i eulers metode også
        if( not(y_iPlussEn[0]>0) ):
            break
        
        y.append(y_iPlussEn)
        xi.append(xi[-1]+h)
    
    return np.array(xi), np.array(y)

#----------------------------------------------------------------
#Hjelpefunksjon
def yToTheta(y):
    theta=y[:,0]
    return theta
#---------------------------------------------------------------------

yInitial=np.array([1,0])

xi1,y1=eulersmethod(yInitial, 0.001,LaneEmden, 1)
theta1=yToTheta(y1)

xi2,y2=eulersmethod(yInitial, 0.1,LaneEmden, 1)
theta2=yToTheta(y2)

xi3,y3=eulersmethod(yInitial, 0.5,LaneEmden, 1)
theta3=yToTheta(y3)

plt.plot(xi1,theta1,label="Numerisk $h=0.001$")
plt.plot(xi2,theta2,label="Numerisk $h=0.1$")
plt.plot(xi3,theta3,label="Numerisk $h=0.5$")
plt.plot(xi1, np.sin(xi1)/xi1, label="Analytisk")
plt.xlabel('$\u03BE$')
plt.ylabel('$\u03B8$')
plt.grid()
plt.legend()
plt.show()
```

Steglengden er valgt til $h=0.001$ siden den numeriske løsningen da har god overenstemmelse med analytisk løsning, samtidig som at koden har kort kjøretid.

```python
def RK4(y0, h, f, n):
    y=[y0]; xi=[10**-3] #starter med liten xi0 siden 0/0 blir tull
    
    while True: 
        k1 = f( y[-1] , xi[-1] , n )
        k2 = f( y[-1]+h*k1/2 , xi[-1]+h/2 , n )
        k3 = f( y[-1]+h*k2/2 , xi[-1]+h/2 , n )
        k4 = f( y[-1]+h*k3 , xi[-1]+h , n )
        S  = (k1 + 2*k2 + 2*k3 + k4)/6
        
        y_iPlussEn=y[-1]+h*S
        
        #y-iPlussEn er siste utregnede theta. Denne er positiv helt til vi er ved enden av planeten.
        #Valgte denne if-løsningen siden RK4 skaper pythonproblemer når theta og chi er veldig små, og unngår dermed
        #feilkode med "nan"
        if( not(y_iPlussEn[0]>0) ):
            break
        
        y.append(y_iPlussEn)
        xi.append(xi[-1]+h)
    
    return np.array(xi), np.array(y)

yInitial=np.array([1,0])

xi1,y1=RK4(yInitial, 0.001,LaneEmden, 1)
theta1=yToTheta(y1)

xi2,y2=RK4(yInitial, 0.1,LaneEmden, 1)
theta2=yToTheta(y2)

xi3,y3=RK4(yInitial, 0.5,LaneEmden, 1)
theta3=yToTheta(y3)

plt.title("RK4 for å finne numerisk løsning av Lane-Emden ligningen")
plt.plot(xi1,theta1,label="Numerisk $h=0.001$")
plt.plot(xi2,theta2,label="Numerisk $h=0.1$")
plt.plot(xi3,theta3,label="Numerisk $h=0.5$")
plt.plot(xi1, np.sin(xi1)/xi1, label="Analytisk")
plt.xlabel('$\u03BE$')
plt.ylabel('$\u03B8$')
plt.grid()
plt.legend()
plt.show()
```

Steglengden er valgt til $h=0.001$ siden den numeriske løsningen da har god overenstemmelse med analytisk løsning, samtidig som at koden har kort kjøretid.

```python
#-------------------------------------------------------------
#Numeriske metoder

def dPdx(P,x,alpha):
    return -0.5*alpha*x * (1+P) * (1+3*P) / (1-x**2*alpha)

def eulersmethod2(P0,h,f,alpha): #Eulers metode spesialtilpasset oppgaven
    xList=np.arange(0,1,h) #x er innhold i [0,1] fra formelen x=r/R
    P=[P0]
    
    for x in xList[1:]: #Ønsker ikke element 0, altså x=0, siden denne inngår i randbetingelsene
        S=f(P[-1],x,alpha)
        P.append(P[-1]+h*S)
    
    return xList,np.array(P)

def RK42(P0,h,f,alpha): #RK4 spesialtilpasset oppgave
    xList=np.arange(0,1,h)
    P=[P0]
    
    for x in list(xList)[1:]: #Ønsker ikke element 0, altså x=0, siden denne inngår i randbetingelsene
        k1 = f( P[-1] , x , alpha )
        k2 = f( P[-1]+h*k1/2 , x+h/2 , alpha )
        k3 = f( P[-1]+h*k2/2 , x+h/2 , alpha )
        k4 = f( P[-1]+h*k3 , x+h , alpha )
        S  = (k1 + 2*k2 + 2*k3 + k4)/6
        P.append(P[-1]+h*S)
    
    return xList,np.array(P)

#------------------------------------------
def analytiskP(x,alpha):
    return (np.sqrt(1-alpha) - np.sqrt(1-alpha*x**2))/(np.sqrt(1-alpha*x**2) - 3*np.sqrt(1-alpha))

def plotter(alpha):
    P0=(np.sqrt(1-alpha) - 1)/(1 - 3*np.sqrt(1-alpha))
    
    xList,P_euler = eulersmethod2(P0,0.001,dPdx,alpha)
    xList,P_RK4 = RK42(P0,0.001,dPdx,alpha)
    
    plt.title("Normaliserte løsninger av ligning (36) ved $\u03B1=$"+str(alpha))
    plt.plot(xList,analytiskP(xList,alpha)/P0,label="Analytisk løsning")
    plt.plot(xList,P_euler/P0,label="Eulers metode")
    plt.plot(xList,P_RK4/P0,label="RK4")
    
    plt.xlabel("$x$")
    plt.ylabel("$P$")
    plt.legend()
    plt.grid()
    plt.show()

#--------------------------------------------

plotter(0.86)
plotter(0.59)
plotter(0.0167)
```

Steglengden er valgt til $h=0.001$ siden de numeriske løsningene da hadde god overenstemmelse med analytisk løsning, samtidig som at koden hadde kort kjøretid.
